# Vector Plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Define the x and y range which corresponds to the list of x and y's that will form our grid.
# In this case, we will have a ~grid of x,y
x_range = y_range =  np.linspace(0,4*np.pi,50)
# Create the pairings of x and y (square grid)
x,y = np.meshgrid(x_range,y_range,)
```

```python
fx, fy = np.sin(x),np.cos(y)
```

### Quiver Plot

[matplotlib.pyplot.quiver](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.quiver.html)

```python
plt.figure(figsize=(6,6))
plt.quiver(x,y,fx,fy,pivot="middle",color='green', alpha=0.7)
plt.axis("scaled") # re-scales y-axis so that it is not squished.
plt.show()
```

### Streamplot

[matplotlib.pyplot.streamplot](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.streamplot.html)

```python
plt.figure(figsize=(6,6))
plt.streamplot(x,y,fx,fy,density=1.5)
plt.axis("scaled")
plt.show()
```
